```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv('/Universities.csv')
```

```python
#plot histogram on SAT column

plt.hist(df['SAT'])
plt.xlabel('SAT')
plt.ylabel('count')
plt.title('Histogram of SAT')
plt.show()
```

```python
sns.histplot(df['Accept'])
plt.xlabel('Accept1')
plt.ylabel('Frequency')
plt.show()
```

```python
# prompt: can you create a histogram on df on sat column can you change the color pink,given the edgecolor to black,can you use seaborn library

import seaborn as sns
import matplotlib.pyplot as plt

# Assuming 'df' is your DataFrame and 'SAT' is the column you want to plot
sns.histplot(df['SAT'], color='pink', edgecolor='black')
plt.xlabel('SAT')
plt.ylabel('Frequency')
plt.title('Histogram of SAT')
plt.show()
```

```python
 plt.plot(df['SAT'], df['Accept'],'rs')
 plt.title('SAT vs Accept')
 plt.xlabel('SAT')
 plt.ylabel('Accept')
 plt.xlabel('SAT')
 plt.ylabel('Accept')
 plt.title('Scatter Plot of SAT vs Accept')
```

```python
# prompt: create  a scatter plot between Top 10 column and accept column and use blue color circle markers in the scatter plot

plt.scatter(df['Top10'], df['Accept'], color='blue', marker='o')
plt.xlabel('Top10')
plt.ylabel('Accept')
plt.title('Scatter Plot of Top10 vs Accept')
plt.show()
```
